```python
import requests
import json
import pickle

## [abo | gun | blm]
campaign = 'blm'

url_user_cascade = 'http://130.56.248.5:9357/active-users/_search?pretty&scroll=1m'
url_user_cascade_scroll = 'http://130.56.248.5:9357/_search/scroll?pretty'

tweets = pickle.load(open('../data/social_media/{}/all_tweets_by_relevant_videos.pkl', 'rb'))
user_ids = [tweets[tweet_id]['_source']['user_id_str'] for tweet_id in tweets]
user_ids = list(set(user_ids))
print(len(user_ids))

query_json = {
    "size": 5000,
    "query": {
        "bool": {
            "should": [
                {"terms": {"user_id_str": user_ids}}
            ]
        }
    }
}

query_str = json.dumps(query_json)

headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}

users = []
scroll_ids = []
total = 0

r = requests.post(url_user_cascade, data=query_str, headers=headers)
result = json.loads(r.content.decode('utf-8'))
hits = result['hits']['hits']
print("# returned hits:", len(hits))
num_total_hits = result['hits']['total']
print("# total hits:", num_total_hits)
print("max_score:", result['hits']['max_score'])
print("=============================================")
last_scroll_id = result['_scroll_id']
scroll_ids.append(last_scroll_id)
total += len(hits)

for hit in hits:
    users.append(hit)

## Scroll search
while len(hits) > 0:
    try:
        scroll_query_json = {
            "scroll" : "1m", 
            "scroll_id" : last_scroll_id
        }
        
        scroll_query_str = json.dumps(scroll_query_json)
        r = requests.post(url_user_cascade_scroll, data=scroll_query_str, headers=headers)
        result = json.loads(r.content.decode('utf-8'))
        hits = result['hits']['hits']
        print("# returned hits:", len(hits))
        total += len(hits)
        print("# total:", total)
        last_scroll_id = result['_scroll_id']
        scroll_ids.append(last_scroll_id)
        for hit in hits:
            users.append(hit)
    except:
        print(result)
        break


print("# users retrieved:", len(users))
print("# total:", total)

results = {}
for user in users:
    userId = user['_source']['user_id_str']
    results[userId] = user

try:
    pickle.dump(results, open("../data/social_media/{}/all_users_by_filtered_videos.pkl".format(campaign), "wb"))
except:
    print("couldn't dump the users!!!")
    pass


## Clear/Close the sessions
clear_scroll_query_json = {"scroll_id": scroll_ids}
clear_scroll_query_str = json.dumps(clear_scroll_query_json)
r = requests.delete(url_user_cascade_scroll, data=clear_scroll_query_str, headers=headers)
```
